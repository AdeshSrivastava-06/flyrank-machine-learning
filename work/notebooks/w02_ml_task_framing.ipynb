## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
Scoring and binary classification. The project ranks content pages according to their likelihood of needing a refresh. A machine learning model estimates the probability that a page belongs to the declining category, and that probability is used to create a prioritized review queue. This approach supports decision-making by helping reviewers focus on the pages with the highest potential impact.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
A fixed rule can only evaluate one or two conditions at a time, such as low CTR or declining traffic. However, the need for a content refresh often depends on multiple interacting factors, including impressions, clicks, CTR, average position, engagement, freshness, and content age. A machine learning model can learn these complex relationships and assign a probability score, producing a more effective ranking than a single if-statement.
```
